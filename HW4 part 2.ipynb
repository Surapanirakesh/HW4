```python
def Distancefn(point, centroid, distType):
     if( distType == "Euclidean"):
            dist = sum((point - centroid)**2)**0.5
            return dist
     elif( distType == "Cosine"):
        A = np.array(point)
        B = np.array(centroid)
        dist = 1 - np.dot(A,B)/(np.linalg.norm(A)*np.linalg.norm(B))
        return dist
     else:
        A = np.array(point)
        B = np.array(centroid)     
        dist = 1 - (np.sum(np.minimum(A,B), axis = 0)/np.sum(np.maximum(A, B), axis = 0)) 
        return dist       
    
def predictfn(classes, y, k =3):
    indexes = []
    for i in range(k):
        indexes.append(np.where(classes == i))
    for cluster in indexes:
        mode = int(stats.mode(y[cluster])[0])
        classes[cluster] = mode
    return classes    
```

```python
def kmeansalg(data,y, k, distType):
    centroids = {}
    starttime = time.time()
    for i in range(k):
        centroids[i] = data[random.randint(i,len(y)-1)]
    classes = {}  
    classlabel = []
    SSE = []
    SSEvalue = 0
    iterations = 0
    while True:
        iterations = iterations + 1
        classes = {}
        classlabel = []
        for i in range(k):
            classes[i] = []
            
        for datapoint in data:
            distance = []
            
            for centroid in centroids:
                dist = Distancefn(datapoint, centroids[centroid], distType)
                distance.append(dist)
                
            minDist = min(distance)
            tempindex = distance.index(minDist)
            flag = 0
            for centroid in centroids:
                if((distance[centroid] == minDist) & (centroid != tempindex)):
                    flag = flag + 1
            if(flag == 0):        
                minindex = tempindex
                
            elif(flag == 1):
                minindex = random.randint(0,1)
            else:
                minindex = random.randint(0,2)    
            classes[minindex].append(datapoint)
            classlabel.append(minindex)
        oldCentroid = dict(centroids)
        
        for key in classes:
            classdata = classes[key]
            if(len(classdata)>0):
                newcentroid = np.mean(classdata, axis = 0)
                centroids[key] = newcentroid
                
        print(centroids) 
        ssevalue = 0
        for key in classes:
            for datapoint in classes[key]:
                ssevalue = ssevalue + Distancefn(datapoint, centroids[key], "Euclidean")
        SSE.append(ssevalue)
#         print(SSE)
        match = False;
        for key in oldCentroid:
            old = oldCentroid[key]
            new = centroids[key]
            if np.sum((new - old)/old * 100) < 0.01:
                match = True
            else: 
                match = False
                break
        if match:
            print("time " + str(time.time()-starttime))
            print("total iterations " + str(iterations) )
            
            print(centroids)
            return centroids, classlabel, classes, SSE 
            break   
```

```python
def label_cluster(cluster, df):
    cl = defaultdict(int)
    for point in cluster:
        cl[point[-1]] += 1
    return cl
```

```python
def alter(cluster, df):
    data = []
    cl = defaultdict(int)
    for point in cluster:
        dftemp = df[(df['sepal_length'] == point[0]) & (df['sepal_width'] == point[1]) & (df['petal_length'] == point[2]) & (df['petal_width'] == point[3])]
        dftemp = dftemp['class'].unique()
        cl[dftemp[-1]] += 1
    print(cl)
    return cl
```
